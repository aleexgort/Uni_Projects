# Forecasting dataset

Da wir Ozon Werte in die Zukunft vorhersagen möchten, müssen wir unser Forecasting Dataset vorbereiten. Das unterstehende Code basiert auf dem folgenden Tutorial: https://www.youtube.com/watch?v=z3ZnOW-S550&t=922s

```python
#We load the necessary packages
from datetime import datetime, timedelta
```

```python
#We create a function to make lags for the other features
def add_lags_features(df, columns):
    for column_name in columns:
        target_map = df[column_name].to_dict()
        df[column_name] = (df.index - pd.Timedelta("48 hours")).map(target_map)
    return df
```

```python
#We have a function to add the time series components
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['hour_of_day'] = df.index.hour
    df['day_of_week'] = df.index.dayofweek + 1
    df['day_of_year'] = df.index.dayofyear
    df['month'] = df.index.month
    df['year'] = df.index.year
    
    
    return df
```

```python
#We create an empty dataset
future = pd.date_range('2023-12-01', '2023-12-02', freq='1h')
future_df_sta = pd.DataFrame(index=future)
future_df_sta['isFuture'] = True #isFuture will appear for the forecasting part
df_sta['isFuture'] = False
df_sta_and_future = pd.concat([df_sta, future_df_sta]) #We concatenate the old and the new dataset
```

```python
df_sta_and_future
```

```python
#We apply our functions to create the lags and the time series features
df_sta_and_future = create_features(df_sta_and_future)
columns_to_lag = ["Luftdruck_STA", "Lufttemp_STA", "Windgeschw_STA", "AnzFahrzeuge_STA"]
add_lags_features(df_sta_and_future, columns_to_lag)
```
